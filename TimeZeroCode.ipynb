```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Read the Excel file
df = pd.read_excel('/Users/janahofmann/Desktop/Thesis/TimeZeroCode/Book2.xlsx')

# Extract x and y values
x = df.iloc[:, 0]  # First column for x-axis
y = df.iloc[:, 1]  # Second column for y-axis

#finding peak for our time-zero
max_index = np.argmax(y)
min_index = np.argmin(y)
x_max = x[min_index]
y_max = y[min_index]
new_x_max = x_max/3
new_y_max = 1-y_max
print(f"Coordinates of the maximum: ({new_x_max}, {new_y_max})")
print(new_x_max)
x_shifted = (x/3) - new_x_max

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_shifted, 1-y, linestyle='-')
plt.xlabel('Probe Delay Time [ps]')
plt.ylabel('Transmittance')
plt.title('Time-Zero of CS2')
plt.grid(True)
plt.savefig('TimeZero.pdf')
plt.show()
```
